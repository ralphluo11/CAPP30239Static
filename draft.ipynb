### DATA CLEAN

```python
import pandas as pd
dem = pd.read_csv('vdem.csv')
dem.head()
```

```python
dem['country_name'].describe()
```

```python
gdp = pd.read_csv('gdp.csv', skiprows=3)
gdp.head() 
```

```python
gdp['Country Name'].describe()
```

```python
year_cols = [col for col in gdp.columns if col.isdigit()]

# melt for long format 
gdp_long = (
    gdp
    .melt(
        id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
        value_vars=year_cols,
        var_name="year",
        value_name="gdp_growth"
    )
    .assign(year=lambda df: df["year"].astype(int))
)

gdp_long.head()
```

```python
dem = dem.rename(columns={"country_text_id": "iso_code"})
gdp = gdp_long.rename(columns={"Country Code": "iso_code"})

df = pd.merge(
    dem,
    gdp,
    on=["iso_code","year"],
    how="left",      
    validate="1:1"   
)
df.head()
```

```python
df = df[df["year"] >= 1970]
df= df.drop(columns=["Country Name", "Indicator Name", "Indicator Code"])
df.head()
```

In this section, I primarily clean and align the two data sources. Although both datasets are well organized, I reshape one of them (using the melt function) to ensure that their formats are consistent and ready for visualization. Since the datasets cover different time periods, I adjust the time range to ensure sufficient overlapping data. Additionally, I drop irrelevant columns to simplify the dataset and improve clarity.

## Overall view of Democracy

```python
yearly = (
    df
    .groupby("year", as_index=False)["v2x_libdem"]
    .mean()
    .rename(columns={"v2x_libdem": "global_libdem"})
)

import altair as alt
# find the min and max for y-axis zoom 
y0 = yearly["global_libdem"].min()
y1 = yearly["global_libdem"].max()

ann = alt.Data(values=[{"year": 1991, "label": "Soviet Union\nCollapses"}, {'year': 2011, 'label': 'Arab Spring'}]) 

base = (
    alt.Chart(yearly)
    .mark_line(point=alt.MarkConfig(size=120, filled=True), interpolate="monotone", strokeWidth=3)
    .encode(
        x=alt.X("year:O", title="Year"),
        y=alt.Y("global_libdem:Q",
                title="Global Avg. Level of Democracy",
                scale=alt.Scale(domain=[y0 - 0.02, y1 + 0.02])),
        tooltip=[
            alt.Tooltip("year:O", title="Year"),
            alt.Tooltip("global_libdem:Q", format=".3f", title="Avg. Level of Democracy")
        ]
    )
)

rule = alt.Chart(ann).mark_rule(color="red", strokeDash=[5,2]).encode(x="year:O")
text = alt.Chart(ann).mark_text(
    dy=30,      
    dx=10,      
    align="left",
    baseline="middle",
    color="red",
    fontWeight="bold"
).encode(
    x="year:O",
    text="label:N"
)

chart = alt.layer(base, rule, text).properties(
    title="Global Level of Democracy (1970–2024)"
)

chart

chart.save("image1.png")
```

I first aim to give viewers a basic understanding of how global democracy has evolved over the years. To provide context, I highlight two influential events — the collapse of the Soviet Union and the Arab Spring — to illustrate their potential impact. The dot and trend graphs accurately depict these changes over time.

```python
# ai help with the list of countries
ex_soviet = [
    "EST","LVA","LTU","BLR","UKR","MDA","GEO","ARM","AZE","KAZ","KGZ",
    "TJK","TKM","UZB","RUS"
]
west_demo = ["GBR","FRA","DEU","BEL","NLD","LUX","CHE","AUT","DNK","NOR","SWE","FIN","USA"]

latin_america = ["ARG","BRA","CHL","COL","MEX","PER","URY"]
```

```python
import numpy as np

def label_region(code):
    if code in ex_soviet:
        return "Ex-Soviet"
    elif code in west_demo:
        return "West Europe + USA"
    elif code in latin_america:
        return "Latin America"
    else:
        return "Rest of the World"

df = df.assign(
    region_group = df["iso_code"].apply(label_region)
)

yearly_region = (
    df[df["year"] >= 1970]
    .groupby(["region_group","year"], as_index=False)["v2x_libdem"]
    .mean()
    .rename(columns={"v2x_libdem":"avg_libdem"})
)
```

```python
import altair as alt

y0 = yearly_region["avg_libdem"].min()
y1 = yearly_region["avg_libdem"].max()

chart = (
    alt.Chart(yearly_region)
    .mark_line(point=True, interpolate="monotone", strokeWidth=3)
    .encode(
        x=alt.X("year:O", title="Year"),
        y=alt.Y("avg_libdem:Q",
                title="Avg. Level of Democracy",
                scale=alt.Scale(domain=[y0-0.02, y1+0.02])),
        color=alt.Color("region_group:N",
                        title="Region Group",
                        scale=alt.Scale(domain=[
                            "Ex-Soviet","West Europe + USA","Latin America","Rest of the World"
                        ],
                        range=["#d62728","#1f77b4","#2ca02c","#7f7f7f"])),
        tooltip=[
            alt.Tooltip("region_group:N", title="Group"),
            alt.Tooltip("year:O", title="Year"),
            alt.Tooltip("avg_libdem:Q", format=".3f", title="Avg Democracy")
        ]
    )
    .properties(
        title="Regional Comparison of Level of Democracy (1970–2024)",
        width=700,
        height=400
    )
)

chart
chart.save("image2.png")
```

After examining the overall global trend and identifying periods of average increase in democracy, it is now important to explore how countries in different regions have performed. ## Freedom of expression and Democracy


```python
import numpy as np

def regime_type(poly):
    if poly >= 0.85:
        return "Full democracy"
    elif poly >= 0.65:
        return "Electoral democracy"
    elif poly >= 0.40:
        return "Electoral autocracy"
    else:
        return "Closed autocracy"

df = df.assign(
    regime = df["v2x_polyarchy"].apply(regime_type)
)
```

```python
from scipy.stats import pearsonr

latest = df[df["year"] == 2024]
r_all, p_all = pearsonr(latest["v2x_freexp_altinf"], latest["v2x_libdem"])

# Base
points = (
    alt.Chart(latest)
    .mark_circle(size=80, opacity=0.8)
    .encode(
        x=alt.X("v2x_freexp_altinf:Q", title="Freedom of Expression"),
        y=alt.Y("v2x_libdem:Q",         title="Level of Democracy"),
        color=alt.Color(
            "regime:N",
            title="Regime Type",
            legend=alt.Legend(title="Regime Type"),
            scale=alt.Scale(
                domain=["Full democracy", "Electoral democracy", "Electoral autocracy", "Closed autocracy"],
                range=["#2ca02c", "#1f77b4", "#ff7f0e", "#d62728"] 
            )
        ),
        tooltip=[
            alt.Tooltip("country_name:N", title="Country"),
            alt.Tooltip("v2x_freexp_altinf:Q", format=".2f", title="Freedom of Expression"),
            alt.Tooltip("v2x_libdem:Q", format=".2f", title="Level of Democracy")
        ]
    )
)

# Reference line
ref_line = (
    alt.Chart(pd.DataFrame({"x":[0,1]}))
    .mark_line(color="black", strokeDash=[5,5])
    .encode(x="x:Q", y="x:Q")
)

# ai helped on how to do the regression 
trend = (
    alt.Chart(latest)
    .transform_loess("v2x_freexp_altinf", "v2x_libdem", groupby=["regime"])
    .mark_line(size=2)
    .encode(
        x="v2x_freexp_altinf:Q",
        y="v2x_libdem:Q",
        color=alt.Color("regime:N", legend=None)
    )
)


text = (
    alt.Chart(pd.DataFrame({
        "x": [0.15], 
        "y": [0.95],
        "label": [f"Overall r = {r_all:.2f}"]
    }))
    .mark_text(fontSize=14, fontWeight="bold")
    .encode(
        x="x:Q",
        y="y:Q",
        text="label:N"
    )
)

final = (
    alt.layer(points, ref_line, trend, text)
       .properties(
           width=700,
           height=500,
           title=alt.TitleParams(
             "2024 Cross-Country Scatter: Expression Freedom vs. Level of Democracy",
             anchor="middle",
             fontSize=18
           )
       )
)
final

final.save("image3.png")
```

This graph examines how the level of democracy relates to freedom of expression — often considered its cornerstone. It also explores whether freedom of expression helps build democracy in less democratic countries or mainly strengthens it in those already democratic.

## GDP and Democracy

```python
panel = df.dropna(subset=["gdp_growth","v2x_libdem"])

# ai helped on the calculation of correlation 
corrs = (
    panel
    .groupby("year")
    .apply(lambda g: pd.Series({
        "r": pearsonr(g["gdp_growth"], g["v2x_libdem"])[0]
    }) if len(g) >= 2 else pd.Series({"r": None}))
    .reset_index()
)


chart = (
    alt.Chart(corrs)
      .mark_line(point=True, strokeWidth=3)
      .encode(
        x=alt.X("year:O", title="Year"),
        y=alt.Y("r:Q", title="Pearson r (GDP growth vs. Democracy)", 
                scale=alt.Scale(domain=[-1,1])),
        tooltip=[
          alt.Tooltip("year:O", title="Year"),
          alt.Tooltip("r:Q", format=".2f", title="Correlation")
        ]
      )
      .properties(
        width=700, height=400,
        title=alt.TitleParams(
          "Yearly GDP Growth vs. Democracy Correlation (1970–2024)",
          anchor="middle", fontSize=18
        )
      )
      .configure_axis(
        labelFontSize=12, titleFontSize=14
      )
)

chart

chart.save("image4.png")
```

## Rule of law 

```python
import pandas as pd
import altair as alt


sample = (
    df[(df.year >= 2014) & (df.year <= 2024)]
      .dropna(subset=["v2x_rule","v2x_libdem"])
      .groupby(["iso_code","country_name","year"])
      .last()
      .reset_index()
      .sample(500, random_state=42)
)

# ai teaches how to bin the data 
sample["rule_bin"] = pd.qcut(
    sample["v2x_rule"],
    5,
    labels=["Lowest","Low","Mid","High","Highest"]
)

bin_stats = (
    sample.groupby("rule_bin")["v2x_rule"]
          .agg(["min","max"])
          .reset_index()
)
bin_stats["rule_mid"] = (bin_stats["min"] + bin_stats["max"]) / 2


agg_bar = (
    sample
      .merge(bin_stats[["rule_bin","rule_mid"]], on="rule_bin")
      .groupby(["rule_bin","rule_mid"])["v2x_libdem"]
      .mean()
      .reset_index(name="avg_libdem")
)


bars = (
    alt.Chart(agg_bar)
      .mark_bar(width=30)
      .encode(
          x=alt.X("rule_mid:Q",
                  title="Rule-of-Law Index (bin midpoint)"),
          y=alt.Y("avg_libdem:Q",
                  title="Avg Liberal Democracy"),
          color=alt.Color("rule_bin:N",
                          title="Rule-of-Law Level",
                          scale=alt.Scale(scheme="blues"))
      )
)


reg_line = (
    alt.Chart(sample)
      .transform_regression("v2x_rule","v2x_libdem", method="linear")
      .mark_line(color="firebrick", size=2)
      .encode(
          x=alt.X("v2x_rule:Q"),
          y=alt.Y("v2x_libdem:Q")
      )
)


final = (
    alt.layer(bars, reg_line)
      .properties(
          width=600, height=400,
          title=alt.TitleParams(
              "Level of Democracy vs. Rule-of-Law",
              subtitle="Bars = avg by bin midpoint; line = continuous regression",
              fontSize=18, anchor="middle"
          )
      )
      .configure_legend(
          orient="right",
          titleFontSize=14,
          labelFontSize=12
      )
)

final

final.save("image8.png")
```

```python
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()

panel = (
    df[(df.year >= 2014) & (df.year <= 2024)]
     .loc[:, [
         "iso_code","country_name","year",
         "v2x_libdem",
         "v2x_freexp_altinf","gdp_growth","v2xeg_eqdr","v2x_rule"
     ]]
     .dropna(subset=["v2x_libdem"])
)

gdp_avg = (
    panel.groupby("iso_code")["gdp_growth"]
         .mean()
         .reset_index(name="gdp")
)
panel = panel.merge(gdp_avg, on="iso_code", how="left")

melted = panel.melt(
    id_vars=["iso_code","country_name","year","v2x_libdem"],
    value_vars=[
      "v2x_freexp_altinf",
      "gdp",
      "v2xeg_eqdr",
      "v2x_rule"
    ],
    var_name="factor",
    value_name="value"
).dropna(subset=["value"])


factor_labels = {
    "v2x_freexp_altinf": "Freedom of Expression",
    "gdp": "Avg GDP Growth",
    "v2xeg_eqdr": "Equal Distribution",
    "v2x_rule": "Rule of Law"
}
melted["factor_label"] = melted["factor"].map(factor_labels)

base = (
    alt.Chart(melted)
      .mark_circle(size=30, opacity=0.4)
      .encode(
          x=alt.X("v2x_libdem:Q", title="Liberal Democracy Index"),
          y=alt.Y("value:Q", title=None),
          color=alt.Color("year:O", title="Year",
                          scale=alt.Scale(scheme="turbo")),
          tooltip=[
            alt.Tooltip("country_name:N", title="Country"),
            alt.Tooltip("year:O",         title="Year"),
            alt.Tooltip("value:Q", format=".2f", title="Value")
          ]
      )
    +
    alt.Chart(melted)
      .transform_loess("v2x_libdem", "value", groupby=["factor_label"])
      .mark_line(color="firebrick", size=2)
      .encode(
          x="v2x_libdem:Q",
          y="value:Q",
          detail="factor_label:N"
      )
    .properties(width=180, height=180)
)


final = (
    base.facet(
        column=alt.Column(
            "factor_label:N",
            title=None,
            header=alt.Header(labelAngle=0, labelFontSize=12)
        )
    )
    .resolve_scale(y="independent")
    .properties(
        title=alt.TitleParams(
            "Core Factors vs. Level of Democracy（2014–2024）",
            fontSize=20, anchor="middle"
        ),
        spacing=10,                  
        padding={"left":20,"right":20,"top":40,"bottom":20}
    )
)

final

final.save("image9.png")
```

```python
# ai teaches how to winsorize and normalize data 
YEAR_MIN, YEAR_MAX = 1990, 2024
FACTOR_COLS = ["v2x_freexp_altinf", "v2xeg_eqdr", "v2x_rule", "gdp_growth"]  
WINSOR_P = 0.01   
LOESS_BW = 0.4    
CHART_W, CHART_H = 620, 420

panel = (
    df[(df.year >= YEAR_MIN) & (df.year <= YEAR_MAX)]
      .loc[:, ["iso_code","country_name","year",
               "v2x_libdem",
               "v2x_freexp_altinf","gdp_growth","v2xeg_eqdr","v2x_rule"]]
      .dropna(subset=["v2x_libdem"])
)

agg = (
    panel.groupby(["iso_code","country_name"])
         .agg(
             v2x_libdem=("v2x_libdem","median"),
             v2x_freexp_altinf=("v2x_freexp_altinf","median"),
             v2xeg_eqdr=("v2xeg_eqdr","median"),
             v2x_rule=("v2x_rule","median"),
             gdp_avg_growth=("gdp_growth","mean")  
         )
         .reset_index()
)

agg = agg.rename(columns={"gdp_avg_growth": "gdp_growth"})

def winsorize_series(s: pd.Series, p=0.01):
    if p is None or p <= 0:
        return s
    lower = s.quantile(p)
    upper = s.quantile(1 - p)
    return s.clip(lower, upper)

def minmax_norm(s: pd.Series):
    mn, mx = s.min(), s.max()
    if np.isclose(mx - mn, 0):
        return pd.Series(0.5, index=s.index)
    return (s - mn) / (mx - mn)

df_a = agg.copy()
for col in FACTOR_COLS:
    x = df_a[col]
    if WINSOR_P is not None:
        x = winsorize_series(x, p=WINSOR_P)
    df_a[f"{col}_norm"] = minmax_norm(x)

df_a["EW_MinMax"] = df_a[[f"{c}_norm" for c in FACTOR_COLS]].mean(axis=1)


def plot_xy(df_in: pd.DataFrame, y_col: str, title: str):
    base = alt.Chart(df_in).encode(
        x=alt.X("v2x_libdem:Q", title="Liberal Democracy Index"),
        y=alt.Y(f"{y_col}:Q",    title=title)
    )

    scatter = base.mark_circle(size=42, opacity=0.45).encode(
        tooltip=[
            alt.Tooltip("country_name:N", title="Country"),
            alt.Tooltip("v2x_libdem:Q",   title="Democracy", format=".2f"),
            alt.Tooltip(f"{y_col}:Q",     title="Composite", format=".2f")
        ]
    )

    trend = (
        base.transform_loess("v2x_libdem", y_col, bandwidth=LOESS_BW)
            .mark_line(size=3)
    )

    return (scatter + trend).properties(width=CHART_W, height=CHART_H)

chart_a = plot_xy(df_a, "EW_MinMax",      "Composite (EW-MinMax, 0–1)")


final = alt.vconcat(
    chart_a.properties(title=alt.TitleParams("All Countries", fontSize=16)),
).resolve_scale(color='independent').properties(
    title=alt.TitleParams(f"Democracy vs. Composite Factors（{YEAR_MIN}–{YEAR_MAX}）",
                          fontSize=18, anchor="middle")
)

final

final.save("image10.png")
```
