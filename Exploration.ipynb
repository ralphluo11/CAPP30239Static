### DATA CLEAN

```python
import pandas as pd
dem = pd.read_csv('vdem.csv')
dem.head()
```

```python
dem['country_name'].describe()
```

```python
gdp = pd.read_csv('gdp.csv', skiprows=3)
gdp.head() 
```

```python
gdp['Country Name'].describe()
```

```python
year_cols = [col for col in gdp.columns if col.isdigit()]

# melt for long format 
gdp_long = (
    gdp
    .melt(
        id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
        value_vars=year_cols,
        var_name="year",
        value_name="gdp_growth"
    )
    .assign(year=lambda df: df["year"].astype(int))
)

gdp_long.head()
```

```python
dem = dem.rename(columns={"country_text_id": "iso_code"})
gdp = gdp_long.rename(columns={"Country Code": "iso_code"})

df = pd.merge(
    dem,
    gdp,
    on=["iso_code","year"],
    how="left",      
    validate="1:1"   
)
df.head()
```

```python
df = df[df["year"] >= 1970]
df= df.drop(columns=["Country Name", "Indicator Name", "Indicator Code"])
df.head()
```

```python
import pandas as pd, altair as alt

pool = (
    df.query("2014 <= year <= 2024")
      .loc[:, ["country_name","v2xeg_eqdr","v2x_libdem"]]
      .dropna()
)

heat = (
    alt.Chart(pool)
      .transform_bin("bx","v2xeg_eqdr", bin=alt.Bin(maxbins=30))
      .transform_bin("by","v2x_libdem",  bin=alt.Bin(maxbins=30))
      .transform_aggregate(count="count()", groupby=["bx","by"])
      .transform_impute(impute="count", key="bx", value=0, groupby=["by"])
      .mark_rect()
      .encode(
          x=alt.X("bx:Q", title="Level of Resource Equality"),
          y=alt.Y("by:Q", title="Level of Democracy"),
          color=alt.Color("count:Q", title="Countries", scale=alt.Scale(scheme="blues")),
          tooltip=["count:Q"]
      )
      .properties(width=650, height=400,
                 title="Where countries sit: Equality vs. Democracy (2D density, 2014â€“2024)")
)

heat
```

```python
# Compute change per country and pick top/bottom movers
ld = df[df.year.isin([2014,2024])][["iso_code","country_name","year","v2x_libdem"]].dropna()
wide = ld.pivot_table(index=["iso_code","country_name"], columns="year", values="v2x_libdem").reset_index()
wide = wide.rename(columns={2014:"ld_2014", 2024:"ld_2024"}).dropna()
wide["delta"] = wide["ld_2024"] - wide["ld_2014"]

top = wide.nlargest(10,"delta")
bot = wide.nsmallest(10,"delta")
sb  = pd.concat([top, bot], axis=0)
sb["label"] = sb["country_name"]

long = sb.melt(id_vars=["label"], value_vars=["ld_2014","ld_2024"], var_name="which", value_name="ldi")
long["year"] = long["which"].map({"ld_2014":2014,"ld_2024":2024})

base = alt.Chart(long).encode(
    x=alt.X("year:O", title=None, axis=alt.Axis(labelAngle=0)),
    y=alt.Y("ldi:Q", title="Liberal Democracy Index"),
    detail="label:N",
    tooltip=["label:N","year:O","ldi:Q"]
)

lines = base.mark_line(color="#374151")
points = base.mark_circle(size=50, color="#111827")

text_left  = (alt.Chart(long[long.year==2014])
                .mark_text(align="right", dx=-6, fontSize=11)
                .encode(y="ldi:Q", x=alt.value(0), text="label"))

text_right = (alt.Chart(long[long.year==2024])
                .mark_text(align="left", dx=6, fontSize=11)
                .encode(y="ldi:Q", x=alt.value(1), text="label"))

(lines + points + text_left + text_right).properties(
    width=650, height=420, title="Largest Changes in Democracy, 2014â†’2024 (Top 10 â†‘ / Top 10 â†“)"
).configure_view(stroke=None)
```

```python
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()


panel = (
    df[(df["year"] >= 2014) & (df["year"] <= 2024)]
      .loc[:, ["iso_code","country_name","year",
               "v2x_libdem", "v2x_freexp_altinf",
               "gdp_growth", "v2xeg_eqdr", "v2x_rule"]]
      .dropna(subset=["v2x_libdem"])
)


gdp_avg = (
    panel.groupby("iso_code")["gdp_growth"]
         .mean()
         .reset_index(name="gdp")
)
panel = panel.merge(gdp_avg, on="iso_code", how="left")

melted = panel.melt(
    id_vars=["iso_code","country_name","year","v2x_libdem"],
    value_vars=["v2x_freexp_altinf", "gdp", "v2xeg_eqdr", "v2x_rule"],
    var_name="factor",
    value_name="value"
).dropna(subset=["value"])


factor_labels = {
    "v2x_freexp_altinf": "Freedom of Expression",
    "gdp":               "Avg GDP Growth",
    "v2xeg_eqdr":        "Equal Distribution",
    "v2x_rule":          "Rule of Law"
}
order = ["Avg GDP Growth", "Equal Distribution", "Freedom of Expression", "Rule of Law"]
melted["factor_label"] = melted["factor"].map(factor_labels)


n_per = 900
sampled = (
    melted.groupby("factor_label", group_keys=False)
          .apply(lambda g: g.sample(n=min(len(g), n_per), random_state=42))
)


points = (
    alt.Chart(sampled)
      .mark_circle(size=22, opacity=0.28, color="#6b7280")  
      .encode(
          x=alt.X("v2x_libdem:Q", title="Liberal Democracy Index"),
          y=alt.Y("value:Q", title=None),
          tooltip=[
              alt.Tooltip("country_name:N", title="Country"),
              alt.Tooltip("year:O",         title="Year"),
              alt.Tooltip("value:Q",        title="Value", format=".2f")
          ]
      )
)

trend = (
    alt.Chart(sampled)
      .transform_loess("v2x_libdem", "value", groupby=["factor_label"])
      .mark_line(color="#b91c1c", size=2.5)                
      .encode(x="v2x_libdem:Q", y="value:Q", detail="factor_label:N")
)

chart = (
    (points + trend)
      .facet(
          column=alt.Column(
              "factor_label:N",
              sort=order, title=None,
              header=alt.Header(labelAngle=0, labelFontSize=12)
          )
      )
      .resolve_scale(x="shared", y="independent")
      .properties(
          title=alt.TitleParams(
              "Core Factors vs. Level of Democracy (2014â€“2024)",
              anchor="middle", fontSize=20
          ),
          spacing=14, padding={"left":20,"right":20,"top":40,"bottom":20}
      )
      .configure_axis(labelFontSize=11, titleFontSize=12)
      .configure_view(stroke=None)
)

chart
```
